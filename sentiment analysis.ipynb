### Review Cleaning

```python
# remove trailing spaces and '\n\n'
def clean(text):
    print(text)
    return text.lstrip('\n\n ').lstrip().rstrip()
    

df['reviews'] = df['reviews'].apply(clean)
```

```python
import string
import nltk
from nltk.corpus import stopwords
from nltk import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
stopwords = stopwords.words('english')
```

```python
def remove_emoji(inputString):
    return inputString.encode('ascii', 'ignore').decode('ascii')
```

```python
import re

def clean_text(text):
#     ps = PorterStemmer()
    lemmatizer = WordNetLemmatizer()
    text = remove_emoji(text)
    text_cleaned = "".join([x for x in text if x not in string.punctuation])
    text_cleaned = re.sub(' +' , ' ', text_cleaned).lower()
    tokens = text_cleaned.split(" ")
    tokens = [token for token in tokens if token not in stopwords]
    text_cleaned = " ".join([lemmatizer.lemmatize(token) for token in tokens])
    
    return text_cleaned
    
```

```python
df['cleaned_reviews'] = df['reviews'].apply(clean_text)
```

### Analysing sentiments using 'vader'

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
def sentiment_analyzer_scores(sentence):
    return analyzer.polarity_scores(sentence)
```

```python
def compound_score(text):
    comp = sentiment_analyzer_scores(text)
    return comp['compound']
```

```python
df['sentiment_score'] = df['reviews'].apply(compound_score)
```

```python
df.sample(5)
```

```python
def sentiment_category(score):
    if score >= 0.05:
        return 'positive'
    elif score <= -0.05:
        return 'negative'
    else:
        return 'neutral'
```

```python
df['review_category'] = df['sentiment_score'].apply(sentiment_category)
```

```python
df.sample(5)
```

### Unigrams

```python
from collections import Counter
```

```python
def get_most_common_words(reviews_list, topn = 20):
    reviews = " ".join(reviews_list)
    tokenized_reviews = reviews.split(" ")
    freq_counter = Counter(tokenized_reviews)
    return freq_counter.most_common(topn)
```

```python
top_20_positive_review_words = get_most_common_words(positive_reviews, 20)
```

```python
top_20_positive_review_words
```

```python
top_20_negative_review_words = get_most_common_words(negative_reviews,20)
top_20_negative_review_words
```

```python
def plot_most_common_words(reviews_list, topn=20, title="Common review words", color="blue", axis = None):
    top_words = get_most_common_words(reviews_list, topn=topn)
    data = pd.DataFrame()
    data['words'] = [val[0] for val in top_words]
    data['freq'] = [val[1] for val in top_words]
    
    if axis!=None:
        sns.barplot(y='words', x='freq', data = data, color=color, ax=axis).set_title(title+" top "+str(topn))
    else:
        sns.barplot(y='words',x='freq',data=data,color=color).set_title(title+" top "+str(topn))
```

```python
from matplotlib import rcParams

rcParams['figure.figsize'] = 8,6

fig,ax = plt.subplots(1,2)
fig.subplots_adjust(wspace=0.5)
plot_most_common_words(positive_reviews, 20, "Positive Review Unigrams", axis = ax[0])
plot_most_common_words(negative_reviews, 20, "Negative Review Unigrams",color='red', axis = ax[1])
```

### Bigrams

Bigrams are a combination of 2 words

Defining a function to create n-grams. It can be a combination of 2,3,4 or more words. This number can be given as an input to the function argument.

```python
def generate_ngram(text,n):
    tokens = text.split(" ")
    ngrams = zip(*[tokens[i:] for i in range(n)])
    return ["_".join(ngram) for ngram in ngrams]
```

```python
positive_review_bigrams = [" ".join(generate_ngram(review, 2)) for review in positive_reviews]
negative_review_bigrams = [" ".join(generate_ngram(review, 2)) for review in negative_reviews]
```

```python
positive_review_bigrams[:5]
```

```python
rcParams['figure.figsize'] = 15,20
fig,ax = plt.subplots(1,2)
fig.subplots_adjust(wspace = 1)
plot_most_common_words(positive_review_bigrams, 40, "Positive Reviews Bigrams", axis = ax[0])
plot_most_common_words(negative_review_bigrams, 40,  "Negative Reviews Bigrams",color="red", axis = ax[1])
```

From the above barplot, we can see that the most common bigrams in positive reviews are good_product, nice_watch, nice_product and those in negative reviews are dont_buy, fake_product, poor_quality.
